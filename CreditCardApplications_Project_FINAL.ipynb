# Here as you can see in the above result that the value occurrences of 0 is much higher than value occurrences of 1.
# This would create a biased ML model if we proceed with the data set at its current state
# Therefore, we are oversampling the minority class i.e 1 using SMOTE and randomly undersampling the majority class i.e 1 to balance the ratio of minority to majority class of this dataset

```python
# separated output column TARGET in a separate variable y and the rest of the columns in X
X=model_data.drop(['TARGET'], axis=1)
y =model_data["TARGET"]

import imblearn
from imblearn import over_sampling, under_sampling
from imblearn.pipeline import make_pipeline

from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import make_pipeline


# Created a pipeline with SMOTE and random undersampler
pipeline = make_pipeline(SMOTE(sampling_strategy=1.0), RandomUnderSampler(sampling_strategy=1.0))

# Fit and transform the data i.2 X and y
X_resampled, y_resampled = pipeline.fit_resample(X, y)


# Split the data into training (80%) and testing (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)

# splitting the test set further into validation and test sets
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state=42)

# Therefore, X_train, y_train are the training sets, X_val, y_val are the validation sets, and X_test, y_test are the test sets
```

```python
# Standardizing the features in the training set using StandardScalar
Scalered=StandardScaler()
Scalered.fit(X_train)
```

```python
# Standardizing the features in the training , validation and testing set using the previously fitted StandardScalar
features = X_train.columns
X_train[features] = Scalered.transform(X_train)
X_val[features] = Scalered.transform(X_val)
X_test[features] = Scalered.transform(X_test)
```

```python
# finding out the columns which are having more than 0.7 correlation with the target column.
def correlation(given_dataset, threshold):
    col_corr = set()
    correlated_matrix = given_dataset.corr()

    for i in range(len(correlated_matrix.columns)):
        for j in range(i):
            if correlated_matrix.iloc[i, j] > threshold:
                col_name = correlated_matrix.columns[i]
                col_corr.add(col_name)
    return col_corr

correlated_features = correlation(X_train[features], 0.7)
print("Correlated Features: %s \n%s" % (len(set(correlated_features)), correlated_features))
```

```python
#This code indicates the standardization of features in the training, validation, and testing sets using a previously fitted StandardScaler.
#Additionally, it identifies and prints the features with a correlation greater than 0.7 with the target column, providing insights into highly correlated features in the dataset.
```

# Evaluating various ML models on the validation set

```python
# Creating a Logistic Regression model
lr = LogisticRegression()

# Fitting the model on the training set
lr.fit(X_train[features], y_train.values.ravel())
```

```python
# predicting the output values and comparing it with the validation set y_val and checking out the model's performance.
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_value_pred=lr.predict(X_val[features])
accuracy=accuracy_score(y_val,y_value_pred)
classification_report=classification_report(y_val,y_value_pred)
confusion_matrix=confusion_matrix(y_val,y_value_pred)
print(accuracy)
print(classification_report)
print(confusion_matrix)
```

```python
#The balanced precision and recall indicate that the model is making reasonably good predictions for both classes.
#However, the presence of false positives and false negatives suggests that there is room for improvement in certain aspects of this logistic regression model's performance.
```

```python
# Creating a Decision Tree Classifier model
dt = DecisionTreeClassifier()
dt.fit(X_train[features], y_train.values.ravel())
```
